```python
# importing modules
import face_recognition
import cv2
import os
from datetime import datetime 

# choice of camera
webcam = cv2.VideoCapture(0)

# loading images
known_image1 = face_recognition.load_image_file("d:Image/Rambha.jpg") # insert name and exstension of desired photo
known_image2 = face_recognition.load_image_file("d:Image/PRAVEEN.jpg") 
known_image3 = face_recognition.load_image_file("d:Image/Avinash.jpg") 
known_image4 = face_recognition.load_image_file("d:Image/Ujjawal.jpg") 
known_image5 = face_recognition.load_image_file("d:Image/Rohit.jpg")
known_image6 = face_recognition.load_image_file("d:Image/Rima.jpg")
known_image7 = face_recognition.load_image_file("d:Image/Phanindra.jpg")

#encoding images
target_encoding1 = face_recognition.face_encodings(known_image1)[0]
target_encoding2 = face_recognition.face_encodings(known_image2)[0]
target_encoding3 = face_recognition.face_encodings(known_image3)[0]
target_encoding4 = face_recognition.face_encodings(known_image4)[0]
target_encoding5 = face_recognition.face_encodings(known_image5)[0]
target_encoding6 = face_recognition.face_encodings(known_image6)[0]
target_encoding7 = face_recognition.face_encodings(known_image7)[0]


process_this_frame = True

while True:
    # capture frame
    ret, frame = webcam.read()
   # gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
   

    # scaling of the acquired frame
    small_frame = cv2.resize(frame, None, fx=0.20, fy=0.20)
    # conversion colors
    rgb_small_frame = cv2.cvtColor(small_frame, 4)

    if process_this_frame:

        # finding and encoding frames
        face_locations = face_recognition.face_locations(rgb_small_frame)
        frame_encodings = face_recognition.face_encodings(rgb_small_frame)

        # matching Photos with camera frames
        if frame_encodings:
            frame_face_encoding = frame_encodings[0]
            match1 = face_recognition.compare_faces([target_encoding1], frame_face_encoding)[0]
            if match1:
                label = "Rambha Devi"
            else:
                match2 = face_recognition.compare_faces([target_encoding2], frame_face_encoding)[0]
                if match2:
                    label = "Praveen Singh" 
                else:
                    match3 = face_recognition.compare_faces([target_encoding3], frame_face_encoding)[0]
                    if match3:
                        label = "Avinash Singh" 
                    else:
                        match4 = face_recognition.compare_faces([target_encoding4], frame_face_encoding)[0]
                        if match4:
                            label = "Ujjawal Singh" 
                        else:
                            match5 = face_recognition.compare_faces([target_encoding5], frame_face_encoding)[0]
                            if match5:
                                label = "Rohit Singh" 
                            else:
                                match6 = face_recognition.compare_faces([target_encoding6], frame_face_encoding)[0]
                                if match6:
                                    label = "Rima Singh" 
                                else:
                                    match7 = face_recognition.compare_faces([target_encoding7], frame_face_encoding)[0]
                                    if match7:
                                        label = "Phanindra Singh"
                                    else: 
                                        label="Sandeep Nayak"

    process_this_frame = not process_this_frame

    # rectangle creation around the face
    if face_locations:
        top, right, bottom, left = face_locations[0]

        top *= 5
        right *= 5
        bottom *= 5
        left *= 5

        cv2.rectangle(frame, (left, top), (right, bottom), (0,255,0), 1) # Green color

        cv2.rectangle(frame, (left, bottom - 25), (right, bottom), (255, 255, 255), cv2.FILLED)  # White color
        label_font = cv2.FONT_HERSHEY_DUPLEX
        cv2.putText(frame, label, (left + 6, bottom - 6), label_font, 0.8, (0, 0, 0), 1) # Black color
     #(...   def Attendance(label):
          #  with open('d:Attendence.csv','w+') as f:
         #       DataList = f.writelines()
        #        nameList = []
       #     for line in DataList:
      #          entry = line.split(',')
     #           nameList.append(entry[0])
    #        if label in nameList:
   #             now = datetime.now()
  #              dtString = now.strftime('%H:%M:%S')
 #               f.writelines(f'\n{label},{dtString}')
 #       Attendance(label) ...) 
    # window name
    cv2.imshow("Face Recognition", frame)

    # stopping capturin loop with "q"
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# releasing and closing all windows
webcam.release()
cv2.destroyAllWindows()
```
